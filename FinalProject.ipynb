### Data Preprocessing

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sn
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/RUDRA COMPUTER/Downloads/ML case Study.csv',header=0)
```

```python
college =pd.read_csv('C:/Users/RUDRA COMPUTER/Downloads/Colleges.csv',header=0)
```

```python
city=pd.read_csv('C:/Users/RUDRA COMPUTER/Downloads/cities.csv',header=0)
```

```python
#Creating lists
Tier1=college['Tier 1'].tolist()
Tier2=college['Tier 2'].tolist()
Tier3=college['Tier 3'].tolist()
```

```python
Tier1
```

```python
#Diffrentiating college tiers
for item in df.College:
    if item in Tier1:
        df['College'].replace(item,1,inplace=True)
    elif item in Tier2:
         df['College'].replace(item,2,inplace=True)
    elif item in Tier3:
         df['College'].replace(item,3,inplace=True)
        
```

```python
df.head()
```

```python
#To know levels of Role
sns.countplot(x="Role",data=df)
```

```python
#Creating lists

Metro=city['Metrio City'].tolist()
Non_Metro=city['non-metro cities'].tolist()
```

```python
for item in df.City:
    if item in Metro:
        df['City'].replace(item,1,inplace=True)
    elif item in Non_Metro:
         df['City'].replace(item,0,inplace=True)
   
```

```python
#Create Dummies variable for 'Role' column

df = pd.get_dummies(df, columns = ['Role'] ,drop_first=True, dtype=int)
```

### Ridge Regression

```python
from sklearn import preprocessing
```

```python
#scaler object store information for x variable
scaler=preprocessing.StandardScaler().fit(X_train)
```

```python
#Standardising independent variables before ridge regression
#transforming X_train ,X_test into X_train and X_test scale
X_train_s=scaler.transform(X_train)
X_test_s=scaler.transform(X_test)
```

```python
#Using Validation curve to find the best value of λ for which R2 value is maximum.
from sklearn.model_selection import validation_curve
```

```python
#this will create 100 values between 10^-2 to 10^8 for λ.
param_range=np.logspace(-2,8,100)
```

```python
#training our model for all λ values
from sklearn.linear_model import Ridge
train_scores,test_scores=validation_curve(Ridge(),X_train_s,y_train,param_name="alpha",param_range=param_range,scoring='r2')
```

```python
train_scores
```

```python
#taking mean of train and test scores
train_mean=np.mean(train_scores,axis=1)
test_mean=np.mean(test_scores,axis=1)
```

```python
#to get best R2
max(test_mean)
```

```python
#To look at R2 value
sns.jointplot(x=np.log(param_range),y=test_mean)
```

```python
#To find λ position (or alpha) of this max R2
np.where (test_mean == max(test_mean))
```

```python
#To get  λ value at 26th index of array
param_range[26]
```

```python
#Fitting model for this best  λ value which have highest R2 value
lm_r_best=Ridge(alpha=param_range[26])
```

```python
lm_r_best.fit(X_train_s,y_train)
```

```python
# Model performance

#R2 values for train and test y variable
r2_score(y_test,lm_r_best.predict(X_test_s))
```

```python
r2_score(y_train,lm_r_best.predict(X_train_s))
```

```python
mean_squared_error(y_train,lm_r_best.predict(X_train_s))
```

```python
r2_score(y_test,lm_r_best.predict(X_test_s))
```

```python
mean_squared_error(y_test,lm_r_best.predict(X_test_s))
```
